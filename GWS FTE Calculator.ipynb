```python
from model_functions import avg_site_size_multiplier, count_activities_subscribed, total_activity_hrs, contract_varibale_mul, delivery_model_multiplier
import pandas as pd
```

```python
raw_data = pd.read_excel('data.xlsx')
```

```python
print(raw_data)
```

```python
df = raw_data
```

```python
# pd.options.display.float_format = '{:.0f}'.format
```

```python
df.info()
```

```python
# Total No. of Sites:
total_no_of_sites = df.groupby('Client')['No. of Sites per country'].sum().reset_index()
```

```python
df = df.merge(total_no_of_sites, on='Client', suffixes=('','_total'))
```

```python
df.rename(columns={'No. of Sites per country_total':'Total No of Sites'}, inplace =True)
```

```python
df
```

```python
df['Average Spend per Site'] = df['Total Managed Spend'] / df['Total No of Sites']
```

```python
df.head()
```

```python
df['Spend per Country'] = df['Average Spend per Site'] * df['No. of Sites per country']
```

```python
df.head()
```

```python
df
```

```python
df['Average Site Multiplier value'] = df.apply(lambda row: avg_site_size_multiplier(row['Region'], row['Average Spend per Site']), axis=1)
```

```python
df
```

```python
df['Adjusted No of Sites'] = df['No. of Sites per country'] * df['Average Site Multiplier value']
```

```python
df.head()
```

```python
Finance_Activity_items = [
    "Billing_Monthly",
    "Fund Reconciliations_Monthly",
    "Payment_Monthly",
    "Monthly JDE period close and Open PO review_Monthly",
    "Monhtly JDE Balance Sheet Reconciliation_Monthly",
    "Corporate Month End Revenue/Costs Adjustments_Monthly",
    "Corporate Pass Through Reconciliation_Monthly",
    "Corporate Payment Entries Submission_Monthly",
    "Internal & External Audit Sampling Requests (SOC1 & SOX)_Monthly",
    "Client JDE Month End Accruals - Open PO, Rental, Utilities etc. - Manual Journal Entries Posting_Monthly",
    "Client Reporting Pack - Standard Report Generation & Distribution_Monthly",
    "Monthly Operational Call for Client Month End Close (Queries from FM - Finance Related)_Monthly",
    "Client Specific reporting (Customised)_Monthly",
    "Corporate P&L Month End Review, Analysis & Queries_Monthly",
    "Corporate Yearly Budget - Preparation, Review & Approval_Monthly",
    "Corporate Monthly Forecast - Preparation, Review & Approval_Monthly",
    "Corporate Operation Merit & Bonus Review_Monthly",
    "Client Budget - Management Fee & Payroll Calculation; Vendor Spend Data_Monthly",
    "Client Forecast - Vendor Spend_Monthly",
    "Client Savings Report_Monthly",
    "Client Finance Audit Supporting_Monthly",
    "Client Billing to Actuals Reconciliation_Monthly"
]
```

```python
df['Activities Subscribed'] = df[Finance_Activity_items].apply(lambda row: row.astype(bool).sum(), axis=1)
```

```python
df
```

```python
df['Total Activity Hours'] = df[Finance_Activity_items].sum(axis=1)
```

```python
df
```

```python
df['Automation Adjustments'] = df['Automation Adjustments'].fillna('None')
```

```python
# Contract variable Multiplier:
df['Contract Structure_mul'] = df.apply(lambda row: contract_varibale_mul("Contract Structure",row['Contract Structure']),axis=1)
df['spend_ratio_mul'] = df.apply(lambda row: contract_varibale_mul("Spend Ratio",row['Fixed/Var. Spend Ratio']),axis=1)
df['principal_mul'] = df.apply(lambda row: contract_varibale_mul("Principal/MA",row['Principal/MA']),axis=1)
df['client_reporting_level_mul'] = df.apply(lambda row: contract_varibale_mul("Client Reporting Level",row['Client Reporting Level']),axis=1)
df['tech_stack_mul'] = df.apply(lambda row: contract_varibale_mul("Technology Stack",row['Tech Stack']),axis=1)
df['automation_adjustment_mul'] = df.apply(lambda row: contract_varibale_mul("Automation Adjustment",row['Automation Adjustments']),axis=1)
df['bank_account_mul'] = df.apply(lambda row: contract_varibale_mul("Bank Account Use",row['Bank Account']),axis=1)
```

```python
df
```

```python
# Multipliers Combo:
df['Contrct_Spend_Auto_CRL_PMA'] = df['Contract Structure_mul'] * df['spend_ratio_mul'] * df['principal_mul'] * df['client_reporting_level_mul'] * df['automation_adjustment_mul']
df['Bank_PMA'] = df['bank_account_mul'] * df['principal_mul']
df['tech_auto_pma'] = df['principal_mul'] * df['tech_stack_mul'] * df['automation_adjustment_mul']
df['Contrct_crl'] = df['Contract Structure_mul'] * df['client_reporting_level_mul']
df['Contrct_crl_auto'] = df['Contract Structure_mul'] * df['client_reporting_level_mul'] * df['automation_adjustment_mul']
df['tech_auto'] = df['tech_stack_mul'] * df['automation_adjustment_mul']
```

```python
df
```

```python
df['adj_time_CSACP'] = df['Billing_Monthly'] * df['Contrct_Spend_Auto_CRL_PMA']
```

```python
df['adj_time_BP'] = (df['Fund Reconciliations_Monthly'] + df['Payment_Monthly']) * df['Bank_PMA']
```

```python
df['adj_time_TAP'] = (df['Monthly JDE period close and Open PO review_Monthly'] + df['Monhtly JDE Balance Sheet Reconciliation_Monthly'] + df['Corporate Month End Revenue/Costs Adjustments_Monthly'] +df['Corporate Pass Through Reconciliation_Monthly'] + df['Corporate Payment Entries Submission_Monthly'] +df['Internal & External Audit Sampling Requests (SOC1 & SOX)_Monthly'] +df['Corporate P&L Month End Review, Analysis & Queries_Monthly'] +df['Corporate Yearly Budget - Preparation, Review & Approval_Monthly']+df['Corporate Monthly Forecast - Preparation, Review & Approval_Monthly']) * df['tech_auto_pma']
```

```python
df['adj_time_CC'] = df['Client JDE Month End Accruals - Open PO, Rental, Utilities etc. - Manual Journal Entries Posting_Monthly'] * df['Contrct_crl']
```

```python
df['adj_time_CCA'] = (df['Client Reporting Pack - Standard Report Generation & Distribution_Monthly']+df['Client Specific reporting (Customised)_Monthly']+df['Client Budget - Management Fee & Payroll Calculation; Vendor Spend Data_Monthly']+df['Client Forecast - Vendor Spend_Monthly']+df['Client Savings Report_Monthly']+df['Client Finance Audit Supporting_Monthly']+df['Client Billing to Actuals Reconciliation_Monthly']) * df['Contrct_crl_auto']
```

```python
df['adj_time_None'] = df['Monthly Operational Call for Client Month End Close (Queries from FM - Finance Related)_Monthly']
```

```python
df['adj_time_TA'] = df['Corporate Operation Merit & Bonus Review_Monthly'] * df['tech_auto']
```

```python
df['Adj_Activity_Hours'] = df['adj_time_CSACP'] + df['adj_time_BP'] + df['adj_time_TAP'] + df['adj_time_CC'] + df['adj_time_CCA'] + df['adj_time_None'] + df['adj_time_TA']
```

```python
df
```

```python
# Add +20% Ad hoc Overlay:
df['after_add_overlay'] = df['Adj_Activity_Hours'] * 1.2
```

```python
df
```

```python
df['adj_time_CSACP_low'] = df['adj_time_CSACP']
df['adj_time_BP_low'] = df['adj_time_BP']
df['adj_time_TAP_low'] = (df['Monthly JDE period close and Open PO review_Monthly'] + df['Monhtly JDE Balance Sheet Reconciliation_Monthly'] + df['Corporate Month End Revenue/Costs Adjustments_Monthly'] +df['Corporate Pass Through Reconciliation_Monthly'] + df['Corporate Payment Entries Submission_Monthly'] +df['Internal & External Audit Sampling Requests (SOC1 & SOX)_Monthly'] +df['Corporate P&L Month End Review, Analysis & Queries_Monthly'] +df['Corporate Monthly Forecast - Preparation, Review & Approval_Monthly']) * df['tech_auto_pma']
df['adj_time_CCA_low'] = (df['Client Forecast - Vendor Spend_Monthly']+df['Client Finance Audit Supporting_Monthly']) * df['Contrct_crl_auto']
```

```python
df
```

```python
df['p'] = (df['adj_time_CSACP_low'] + df['adj_time_BP_low'] + df['adj_time_TAP_low'] + df['adj_time_CCA_low']) / df['Adj_Activity_Hours']
```

```python
df
```

```python
df['adj_time_after_dm'] = df.apply(lambda row: delivery_model_multiplier(row['Delivery Model'],row['after_add_overlay'],row['p']),axis=1)
```

```python
df
```

```python
# Save to excel file:
df.to_excel('output_data.xlsx', index=False)
```
